```python
sns.relplot(x="city_mpg", y="highway_mpg", data = df)
```

```python
sns.relplot(x="city_mpg", y="highway_mpg", hue = "fuel_system", data = df)
```

```python
sns.relplot(x="bore", y="stroke", hue = "engine_location", data = df);
```

```python
sns.relplot(x="curb_weight", y="engine_size", size = "engine_type", data = df);
```

```python
sns.catplot(x="peak_rpm", y="price", data=df)
```

```python
sns.catplot(x="peak_rpm", y="price", jitter = False,  data=df)
```

```python
sns.catplot(x="city_mpg", y="highway_mpg", hue = "fuel_system", data=df)
```

```python
sns.catplot(x="peak_rpm", y="price", kind = "swarm", data=df)
```

```python
sns.catplot(x="city_mpg", y="highway_mpg", hue = "fuel_system", kind = "swarm", data=df)
```

```python
sns.catplot(x="peak_rpm", y="price", kind = "box", data=df)
```

```python
sns.catplot(x="peak_rpm", y="price", hue = "engine_location", kind = "box", data=df)
```

```python
sns.catplot(x="peak_rpm", y="price", hue = "engine_location", kind = "violin", data=df)
```

```python
sns.catplot(x="peak_rpm", y="price", hue = "engine_location", kind = "violin", split = True, data=df)
```

```python
sns.catplot(x="peak_rpm", y="city_mpg", hue = "fuel_system",  kind = "bar", data=df)
```

```python
sns.catplot(x="peak_rpm", y="city_mpg", hue = "fuel_system", kind = "point", data=df)
```

```python
sns.catplot(x="peak_rpm", y="city_mpg", hue="fuel_system",
            col="engine_location", aspect=.9,
            kind="swarm", data=df);
```
